Which advertising channel has a strong relationship with sales volume, and can be used to model and predict the sales. 

### Step 1: Read the dataset and inspect its columns and 5-point statistics

```python
# Load necessary libraries and import the data
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('seaborn')

data = pd.read_csv('Advertising.csv', index_col=0)
```

```python
# Check the columns and first few rows
data.head()
```

```python
# Get the 5-point statistics for data 
data.describe()
```

```python
# Describe the contents of this dataset
# Each row contains values for how much a campaign spent on addspace in three areas, TV, radio, and newspaper.
# According to the means, campaigns spent the most on TV and the least on radio ads.  There are 200 campaigns sampled.
# There is also a target variable sales showing the sales records of each campaign.
```

### Step 2: Plot histograms with kde overlay to check for the normality of the predictors

```python
# For all the variables, check if they hold normality assumption
for col in data:
    data[col].plot.hist(density=True, label=col)
    data[col].plot.kde(label="KDE")
    plt.ylabel('Density')
    plt.legend()
    plt.show()
```

```python
# Record your observations on normality here 
# The TV and radio variables have relatively normal symetrical distributions.
# The newspaper variable is less symetrical because it has a moderate positive skew, but is still generally normal.
# The sales target variable has a normal distribution with a gentle positive skew.
```

### Step 3: Test for the linearity assumption. 
Use scatterplots to plot each predictor against the target variable

```python
# visualize the relationship between the preditors and the target using scatterplots
scat_fig = plt.figure(figsize = (14,6))
TV_scat = scat_fig.add_subplot(131)
TV_scat.scatter(data.TV, data.sales, label='TV')
TV_scat.set_xlabel('TV')
TV_scat.set_ylabel('sales')
TV_scat.legend()

radio_scat = scat_fig.add_subplot(132)
radio_scat.scatter(data.radio, data.sales, label='radio')
radio_scat.set_xlabel('radio')
radio_scat.set_ylabel('sales')
radio_scat.legend()

new_scat = scat_fig.add_subplot(133)
new_scat.scatter(data.newspaper, data.sales, label='newspaper')
new_scat.set_xlabel('newspaper')
new_scat.set_ylabel('sales')
new_scat.legend()
```

```python
# Record yor observations on linearity here 
# TV has a linear relationship with sales and a regression should be found.
# Radio has a slight linear relationship with sales and is worth trying a regression.
# Newspaper seems to have no linear relationship with sales.
```

### Conclusion so far !

Based on above initial checks, we can confidently say that TV and radio appear to be good predictors for our regression analysis. Newspaper is very heavily skewed and also doesnt show any clear linear relationship with the target. 
> We shall move ahead with our analysis using TV and radio , and count out the newspaper due to the fact that data violates ols assumptions

Note: Kurtosis can be dealt with using techniques like log normalization to "push" the peak towards the center of distribution. We shall talk about this in the next section. 

### Step 4: Run a simple regression in `statsmodels` with TV as a predictor

```python
# import libraries
# already done!
# build the formula 
f = 'sales~TV'
# create a fitted model in one line
model = ols(formula=f, data=data).fit()
```

### Step 5: Get regression diagnostics summary


```python
model.summary()
```

### Record your observations on "Goodness of fit"

The model explains about 61% of the variations in sales.  The amount spent on TV adds has a moderate correlation (not causation) on sales.


Note here that the coefficients represent associations, not causations


### Step 6:  Draw a prediction line with data points omn a scatter plot for X (TV) and Y (Sales)

Hint: We can use `model.predict()` functions to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
TV_new = data[(data.TV == data.TV.min()) | (data.TV ==data.TV.max())]
TV_new = TV_new.drop(['radio','newspaper','sales'], axis=1)
print(TV_new.head())
# make predictions for those x values and store them
predictions = model.predict(TV_new)
print(predictions)
# first, plot the observed data and the least squares line
plt.scatter(data.TV, data.sales)
plt.plot(TV_new, predictions, color='red')
plt.xlabel('TV')
plt.ylabel('sales')
```

### Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig1 = plt.figure(figsize =(15,8))
fig1 = sm.graphics.plot_regress_exog(model, 'TV', fig=fig1)
```

```python
# Record Your observations on residuals
# There is a lot of heteroscedasticity in the first two charts. The data creates a cone as the TV variable increases.
```

### Next, repeat above for radio and go through the same process, recording your observations

```python
f2 = 'sales~radio'
model2 = ols(formula=f2, data=data).fit()

radio_new = data[(data.radio == data.radio.min()) | (data.radio ==data.radio.max())]
radio_new = radio_new.drop(['TV','newspaper','sales'], axis=1)
print(radio_new)
predictions2 = model2.predict(radio_new)
print(predictions)

plt.scatter(data.radio, data.sales)
plt.plot(radio_new, predictions2, color='red')
plt.xlabel('radio')
plt.ylabel('sales')

fig2 = plt.figure(figsize=(15,8))
fig2 = sm.graphics.plot_regress_exog(model2, 'radio', fig=fig2)
```

```python
#There is again a fair amount of heteroscedasticity. The data creates a cone as the radio variable increases.
```

```python
model2.summary()
```

```python
# Record your observations here for goodnes of fit 
# With a R-squared of .332, the model only explains 33% of the variance in sales, which is very low.
# There is also higher skew and kurtosis.
# Radio seems to be a worse predictor of sales than TV.
```
